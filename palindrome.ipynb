## palindrome checking


```python
def palindrome(n):
    if n < 0:
        return False
    original=n
    reverse=0
    while n >0 :
        digit= n % 10
        reverse=reverse * 10 +digit
        n =n // 10
    return original==reverse
```

```python
print(palindrome(123))
```

```python
print(palindrome(121))
```

```python
def plaindrome1(n):
    return str(n)== str(n)[::-1]
```

```python
print(plaindrome1(123))
```

```python
print(plaindrome1(555))
```
