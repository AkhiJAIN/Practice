## using common technique

```python
def prefix(str):
    if not str:
        return ""
    prefix=str[0]

    for word in str[1:]:
        while word[:len(prefix)]!=prefix:
            prefix=prefix[:-1]
            if not prefix :
                return ""
    return prefix
```

```python
input=['aabb','aacc','aaa']
```

```python
print(prefix(input))
```

## usning the brute force 

```python
def prefixmost(str):
    if not str:
        return ""
    result=""

    for i in range(len(str[0])):
        char=str[0][i]
        for word in str:
            if i >= len(word) or word[i]!=char:
                return result
        result += char

    return result
```

```python
input=['aabb','aacc','aaa']
```

```python
print(prefixmost(input))
```
