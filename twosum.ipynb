## using the hash map

def twosum(nums,target)

```python
def twosum(nums,target):
    hashmap={}

    for i in range(len(nums)):
        complment=target-nums[i]

        if complment in hashmap:
            return(hashmap[complment],i)
        hashmap[nums[i]]=i
        
```

```python
nums=[2,7,11,12,13]
target=9
```

```python
print(twosum(nums,target))
```

## using two pointers

```python
def twosum(nums,target):
    left=0
    right=len(nums)-1

    while left<right:
        total=nums[left]+nums[right]

        if total==target:
            return (left,right)
        elif total < target:
            left=+1
        else:
            right-=1
```

```python
nums=[1,2,3,4,5]
target=5
print(twosum(nums,target))
```
